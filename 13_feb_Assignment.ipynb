```python
#Problem_3
"""
exception ArithmeticError
ArithmeticError is simply an error that occurs during numeric calculations.
This class is the base class for those built-in exceptions that are raised for various arithmetic errors such as :
OverflowError
ZeroDivisionError
FloatingPointError
"""
#Example 

try:
	a = 10/0
	print (a)
except ArithmeticError:
		print ("This statement is raising an arithmetic exception.")
else:
	print ("Success.")


#example

j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
#Problem_4
"""
The LookupError exception in Python forms the base class for all exceptions that are raised when an index 
r a key is not found for a sequence or dictionary respectively.

LookupError excep
tion class to handle both IndexError and KeyError exception classes.

"""
#Example 1 - Handling IndexError exception
# lists
x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")

#Example 2 - Handling KeyError exception
in_info = {'name': 'Lenin Mishra',
                'age': 28,
                'language': 'Python'}
user_input = input('What do you want to learn about Pylenin==> ')

try:
    print(f'{user_input} is {in_info[user_input]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```
